```python
class GridCell:
    """ This class holds the coordinates - start and end of latitude and longitude """
    def __init__(self, llat, ulat, llon, ulon, fire_count):
        self.llat = llat
        self.ulat = ulat
        self.llon = llon
        self.ulon = ulon
        self.fire_count = fire_count
        
    def __repr__(self):
        return f"GridCell( llat = {self.llat},  ulat = {self.ulat}, llon = {self.llon}, ulon = {self.ulon}, fire_count = {self.fire_count})"

```

```python
meridians_length = len(meridians) # x-axis - longitude
parallels_length = len(parallels) # y-axis - latitude

grid =  np.full(shape = (meridian_length,parallels_length), fill_value = None,)

meridians_list = list(meridians)
parrallels_list = list(parallels)

for row in range(0, meridians_length):
    for col in range(0, parallels_length):
        latitude = parrallels_list[col]
        longitude = meridians_list[row]
        lower_latitide = latitude
        upper_latitude = latitude + 1
        lower_longitude = longitude
        upper_longitude = longitude + 1
        cell = GridCell(lower_latitide, upper_latitude, lower_longitude, upper_longitude, 0)
        grid[row, col] = cell

grid
```
